```python

def compute_similarity(fix1, fix2):
    spatial_distance = np.sqrt((fix1[0] - fix2[0])**2 + (fix1[1] - fix2[1])**2)
    temporal_difference = abs(fix1[2] - fix2[2])  # Absolute difference in fixation durations    
    similarity = -spatial_distance - temporal_difference
    return similarity

def needleman_wunsch_scanpaths(scanpath1, scanpath2):

    n = len(scanpath1)
    m = len(scanpath2)
    
    similarity_matrix = np.zeros((n+1, m+1))    
    traceback_matrix = np.zeros((n+1, m+1), dtype=int)    
    gap_penalty = -60
    
    # Fill the similarity matrix
    for i in range(1, n+1):
        for j in range(1, m+1):
            # Calculate the scores for match, insertion, and deletion
            match_score = similarity_matrix[i-1, j-1] + compute_similarity(scanpath1[i-1], scanpath2[j-1])
            delete_score = similarity_matrix[i-1, j] + gap_penalty
            insert_score = similarity_matrix[i, j-1] + gap_penalty
            
            # Determine the maximum score among match, insertion, and deletion
            max_score = max(match_score, delete_score, insert_score)
            similarity_matrix[i, j] = max_score
            
            # Update the traceback matrix to keep track of the direction
            if max_score == match_score:
                traceback_matrix[i, j] = 0  # Match
            elif max_score == delete_score:
                traceback_matrix[i, j] = 1  # Delete
            else:
                traceback_matrix[i, j] = 2  # Insert
    
    # Backtrack to find the aligned sequences
    aligned_scanpath1 = []
    aligned_scanpath2 = []
    i, j = n, m
    while i > 0 or j > 0:
        if traceback_matrix[i, j] == 0: 
            aligned_scanpath1.append(scanpath1[i-1])
            aligned_scanpath2.append(scanpath2[j-1])
            i -= 1
            j -= 1
        elif traceback_matrix[i, j] == 1:  # Delete or upward movement
            aligned_scanpath1.append(scanpath1[i-1])
            aligned_scanpath2.append((0, 0, 0))  # Represent gap with zeros
            i -= 1
        else:  # Insert or leftward movement
            aligned_scanpath1.append((0, 0, 0))  # Represent gap with zeros
            aligned_scanpath2.append(scanpath2[j-1])
            j -= 1
    
    # Reverse the aligned sequences to get the correct order
    aligned_scanpath1.reverse()
    aligned_scanpath2.reverse()
    
    return aligned_scanpath1, aligned_scanpath2

# Example usage:
scanpath1 = [(100, 200, 300), (150, 250, 350), (200, 300, 400)]  # Example scanpath 1
scanpath2 = [(110, 210, 320), (160, 260, 340), (220, 310, 410)]  # Example scanpath 2

aligned_scanpath1, aligned_scanpath2 = needleman_wunsch_scanpaths(scanpath1, scanpath2)
print("Aligned Scanpath 1:", aligned_scanpath1)
print("Aligned Scanpath 2:", aligned_scanpath2)
```
