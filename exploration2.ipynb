```python
import pandas as pd
import os 
import scanpath
import math
import numpy as np
```

```python
def get_file_data(file):
    file_path = os.path.join('./data/scanpaths/', file)
    df = pd.read_csv(file_path, delimiter='\t')
    cols = df.columns.to_list()
    df['file'] = file.replace("_scanpath.tsv", "")
    df = df[['file'] + cols]
    return df
```

```python
files = [i for i in os.listdir('./data/scanpaths/')]
files = [i for i in files if ('reader85_p0' in i)|('reader72_p0' in i)]
master_df = get_file_data(files[0])

for i  in  files[1:]:
    master_df = pd.concat([master_df, get_file_data(i)])
```

```python
master_df = master_df.rename(columns={'fixation_position_x':'x', 'fixation_position_y':'y', 'fixation_duration':'duration'})
s = master_df[master_df['file'] == 'reader85_p0'][['x', 'y', 'duration']]
t = master_df[master_df['file'] == 'reader72_p0'][['x', 'y', 'duration']]
```

```python
center_x = master_df['x'].median()
center_y = master_df['y'].median()

distance = math.sqrt(pow(center_x,2) + pow(center_y,2))
score, path, alignment  = scanpath.rscasim(s, t, center_x, center_y, distance, 1, modulator=0.83)
```
